### Question 1: Monthly Cycle for All Years

Create a plot that shows the monthly cycle for all the years in the dataset. 

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from pathlib import Path

# Small style adjustments for more readable plots
plt.style.use("seaborn-whitegrid")
plt.rcParams["figure.figsize"] = (8, 6)
plt.rcParams["font.size"] = 14
```

```python
DATA_DIR = Path.home()/Path('shared/climate-data')

co2_file = DATA_DIR / "monthly_in_situ_co2_mlo_cleaned.csv"
data = pd.read_csv(co2_file, na_values=-99.99).dropna()
```

```python
sns.lineplot(data=data, x='month', y='c02', hue='year')
plt.legend(bbox_to_anchor = (1.05, 1)) #adjsut the group labels position
plt.show()

plt.savefig('outputs/hw3_Q01b.png')
```

According to above plot, we found that the overall CO2 in Manua Loa increase from 1965 to 2010. For each year, CO2 level started to increase and got highest in May. After May, CO2 started decrease and got lowest in September.
