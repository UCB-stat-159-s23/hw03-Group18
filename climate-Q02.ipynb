```python
# using a different method
dy = data.groupby('year')[['c02']].mean()
merged = data.merge(dy, how='inner', on='year') 
merged['c02'] = merged['c02_x'] - merged ['c02_y']
monthly_anomaly = merged.groupby('month').mean()[['c02']]
monthly_anomaly
```
